```python
import librosa
import numpy as np
from IPython.display import Audio
import pandas as pd

def augment_audio(file_path, mode="change pitch", factor=None):

    data, sr = librosa.load(file_path, sr=None)

    if mode == "change pitch":
        if factor is None:
            factor=0.005
        augmented = librosa.effects.pitch_shift(data, sr=sr, n_steps=factor)
    elif mode == "noise":
        if factor is None:
            factor=1.022
        noise = np.random.randn(len(data))
        augmented = data + factor * noise
        augmented = augmented.astype(data.dtype)
    else:
        augmented = data

    return Audio(augmented, rate=sr)
```

```python
X_train = pd.read_csv("csvki/asv2019_inthewild_dataset.csv")
```

```python
singel_data = X_train.iloc[567]
```

```python
augment_audio(singel_data['filepath'], "noise", factor=0.005)
```

```python
augment_audio(singel_data['filepath'], "speed", factor=1.025)
```
